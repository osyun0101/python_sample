```python
import matplotlib.pyplot as plt
import cv2
from scipy import ndimage

cascade_file = "haarcascade_frontalface_alt.xml"
cascade = cv2.CascadeClassifier(cascade_file)

img = cv2.imread("kanzi.jpg")


def face_detect(im):
    img_g = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    im_c = cascade.detectMultiScale(img_g, minSize=(300, 300))
    for (x,y,w,h) in im_c:
        red = (0,0,255)
        cv2.rectangle(im, (x,y),(x+w, y+h),red, thickness=30)
    
for i in range(0, 9):
    turn = i * 10
    nd = ndimage.rotate(img, turn)
    face_detect(nd)
    plt.subplot(3,3,i+1)
    plt.axis("off")
    plt.title("kanzi="+ str(turn))
    plt.imshow(cv2.cvtColor(nd, cv2.COLOR_BGR2RGB))

plt.show
```
