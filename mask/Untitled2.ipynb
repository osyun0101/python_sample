```python
import cv2, os, dlib, pprint
from keras.models import load_model

red =(0,0, 255)
green = (0, 255, 0)
fid = 1
detector = dlib.get_frontal_face_detector()
labels = ['ok', 'MASK_OFF!!']

cap = cv2.VideoCapture(0)
while True:
    ok, frame = cap.read()
    frame = cv2.resize(frame, (500, 300))
    if not ok: break
    dec = detector(frame, 1)
    pprint.pprint(dec)
    for d in dec:
        x1 = int(d.left())
        y1 = int(d.top())
        x2 = int(d.right())
        y2 = int(d.bottom())
        im = frame[y1:y2, x1:x2]
        im = cv2.resize(im, (50,50))
        im = im.reshape(-1, 50,50,3)
        model = load_model('mask-model.hdf5')
        pre = model.predict(im)[0]
        m = pre.argmax()
        print(labels[m], pre[m])
        
        if m == 1:
            color = red
            border = 7
        else:
            color = green
            border = 2
        cv2.rectangle(frame, (x1,y1),(x2,y2),color,thickness=border)
        cv2.putText(frame, labels[m], (x1, y1-7), cv2.FONT_HERSHEY_SIMPLEX, 0.9,color, thickness=border)
        if len(dec) > 0:
            path = './'+'live'+'/'+str(fid)+'.png'
            if not os.path.exists(path):
                cv2.imwrite(path, frame)
                fid += 1
        cv2.imshow('LIVE_MASK', frame)
        
        
    k = cv2.waitKey(1)
    if k == 27 or k == 13:break
        
cap.release()
cv2.destroyAllWindows()
nekoneko
```
