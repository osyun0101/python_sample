```python
import dlib, os, glob,pprint
import cv2

indir = './capture'
outdir = './face'
fid = 1000
detector = dlib.get_frontal_face_detector()

def get_dec(path):
    global fid
    img = cv2.imread(path)
    ltrb = detector(img,1)
    for d in ltrb:
        pprint.pprint(d)
        x1 = int(d.left())
        y1 = int(d.top())
        x2 = int(d.right())
        y2 = int(d.bottom())
        im = img[y1:y2, x1:x2]
        try:
            im = cv2.resize(im, (50,50))
        except:
            continue
        savepath = outdir+'/'+str(fid)+'.jpg'
        cv2.imwrite(savepath, im)
        fid +=1
        
        
files = glob.glob(indir+'/*')
for f in files:
    print(f)
    get_dec(f)
print('ok')
```
