```python
from keras.utils import to_categorical
from keras.models import load_model
import sqlite3
import numpy as np
import os

path = './hw.sqlite3'
weight_path = './hw-mode-weight.h5'
sql ="SELECT * FROM person ORDER BY id DESC LIMIT 5000"

x=[]
y=[]
with sqlite3.connect(path) as conn:
    for row in conn.execute(sql):
        id, height, weight, type_no=row
        height = height / 200
        weight = weight / 150
        x.append(np.array([height, weight]))
        y.append(type_no)
        
y = to_categorical(y, 6)

model = load_model('hw-model.h5')

if  os.path.exists(weight_path): 
    model.load_weights(weight_path)
    
model.fit(np.array(x), y, batch_size=50, epochs=100)
model.save_weights(weight_path)
```
