```python
from detect_zip import *
import matplotlib.pyplot as plt
import pickle

with open('already.data', "rb") as fp:
    clf = pickle.load(fp)
    
cnts, img = detect_zipno("hagaki1.png")

for i,cn in enumerate(cnts):
    x,y,w,h = cn
    x += 8
    y += 8
    w -= 16
    h -= 16
    im2 = img[y:y+h, x:x+w]
    im = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
    im = cv2.resize(im, (8,8))
    im = 15 - im//16
    im = im.reshape(-1, 64)
    
    pre = clf.predict(im)
    
    plt.subplot(1,7,i+1)
    plt.imshow(im2)
    plt.axis("off")
    plt.title(str(pre))

plt.show()
```
