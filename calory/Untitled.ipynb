```python
from flickrapi import FlickrAPI
import os,time
from pprint import pprint
from urllib.request import urlretrieve


key = 'b389b9fbfb5e3e8e94727972a8110f29'
secret = 'bd3234eb67d34088'
wait = 1
im_path = './image'

def main():
    go_download('マグロ 寿司', 'sushi')
    go_download('サラダ', 'salad')
    go_download('麻婆豆腐', 'tohu')
    
def go_download(keyword, dir):
    path = im_path + '/'+ dir
    if not os.path.exists(path):
        os.mkdir(path)
    flickr = FlickrAPI(key, secret, format='parsed-json')
    res = flickr.photos.search(
        text = keyword,
        per_page= 300,
        media = 'photos',
        sort = 'relevance',
        safe_search = 1,
        extras = 'url_q, license'
    )
    photos = res['photos']
    pprint(photos)
    
    try:
        for i,v in enumerate(photos['photo']):
            filepath = path + '/'+v['id']+'.jpg'
            if os.path.exists(filepath): continue
            print(str(i+1)+'download=', v['url_q'])
            urlretrieve(v['url_q'], filepath)
            time.sleep(wait)
    except:
        import traceback
        traceback.print_exc()

        

if __name__=='__main__':
    main()
```
