```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

path ='./image/photos.npz'
data = np.load(path)
photos = data['x']
photo = photos[12]

plt.
for i in range(36):
    plt.subplot(6,6,i+1)
    center = (16, 16)
    scale = 1.0
    roll = i*5
    img = cv2.getRotationMatrix2D(center, roll, scale)
    img = cv2.warpAffine(photo, img, (32, 32))
    plt.imshow(img)
    
plt.show()
```
