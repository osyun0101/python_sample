```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

photos = np.load('image/photos.npz')
x = photos['x']
img = x[12]

plt.figure(figsize=(10, 10))
for i in range(36):
    plt.subplot(6, 6, i+1)
    center = (16, 16)
    angle = i * 5
    scale = 1.0
    mtx = cv2.getRotationMatrix2D(center, angle, scale)
    img2 = cv2.warpAffine(img, mtx, (32, 32))
    plt.imshow(img2)
    
plt.show()
```
