```python
import cv2
import matplotlib.pyplot as plt

kanzi = cv2.imread("kanzi.jpg")
kanzi = cv2.cvtColor(kanzi, cv2.COLOR_BGR2GRAY)

im = cv2.GaussianBlur(kanzi,  (7,7),0)
im = cv2.threshold(im, 140, 255,cv2.THRESH_BINARY_INV)[1]

cnts= cv2.findContours(im, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[1]

for pt in cnts:
    x,y,w,h = cv2.boundingRect(pt)
    if w < 30 or w >300: continue
    cv2.rectangle(im, (x,y),(x+w,y+w),(0,0,255), 3)
    

plt.imshow(im)
plt.show()
```
