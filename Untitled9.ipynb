```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("tana.jpg")
h,w = img.shape[:2]

img = img[0:h, w//3:w]

img_g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_g = cv2.GaussianBlur(img_g, (1,1),0)
img_g = cv2.threshold(img_g, 140, 255, cv2.THRESH_BINARY_INV)[1]

plt.imshow(img_g)
```
