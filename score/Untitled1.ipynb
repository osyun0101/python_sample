```python
import pickle
import os, glob
import matplotlib.pyplot as plt
import cv2

outdir = './png_file'
plt.figure(figsize=(9,17))
n = list(range(177,221))
n.append(166)
n.append(221)
im_size = 25
save_file = outdir+'/katakana.pickle'

result = []
for i, d in enumerate(n):
    img_dir = outdir+"/"+str(d)
    di = glob.glob(img_dir+"/*")
    print(img_dir)
    for j, k in enumerate(di):
        im = cv2.imread(k)
        img_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        im = cv2.resize(img_gray, (im_size, im_size))
        result.append([i, im])
        if j == 3:
            plt.subplot(11, 5, i+1)
            plt.title(str(i))
            plt.axis('off')
            plt.imshow(im, cmap='gray')

pickle.dump(result, open(save_file, 'wb'))
plt.show()
print("ok")
```
